```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import norm

# Datos Medidos: Central values extracted from the table
central_values = np.array([
    3.0, 2.8, 2.6, 2.8, 2.7, 2.7, 2.7, 2.7, 2.6, 2.7,
    2.7, 2.8, 2.8, 2.7, 2.8, 3.0, 3.0, 2.7, 2.7, 3.0
])

# Midpoints and their corresponding frequencies
midpoints = [2.6, 2.7, 2.8, 3.0]
counts = [2, 9, 5, 4]

# Create a bar chart for the measurements
plt.bar(midpoints, counts, width=0.1, align='center',
        edgecolor='black', alpha=0.7, label='Mediciones')

# Calculate mean and standard deviation of the data
mu = np.mean(central_values)
sigma = np.std(central_values, ddof=1)  # Sample standard deviation

# Generate x values for the normal distribution curve
x = np.linspace(2.5, 3.1, 1000)

# Calculate scaled normal distribution (scaled by total count * bin width)
scaling_factor = len(central_values) * 0.2  # 20 * 0.2 = 4
normal_curve = norm.pdf(x, mu, sigma) * scaling_factor

# Plot the normal distribution
plt.plot(x, normal_curve, 'r-', linewidth=2,
         label='Distribución Normal Ajustada')

# Add labels and title
plt.xlabel('Resistencia (Ω)')
plt.ylabel('Frecuencia de dato')
plt.title('Distribución de Mediciones de Resistencia con Incertidumbre')
plt.legend()
plt.grid(True, linestyle='--', alpha=0.7)

# Show the plot
plt.show()
```
