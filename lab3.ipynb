# Linear Regression

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import linregress
```

```python
# Experimental Data here:

v = np.array([4.08, 3.548, 2.952, 2.203, 1.845, 1.152, 0.692, 3.051, 2.474, 1.35])
i = np.array([0.417, 0.362, 0.300, 0.221, 0.186, 0.115, 0.069, 0.307, 0.249, 0.135])

v=np.sort(v)
i=np.sort(i)

print(v)
print(i)

if v.size == i.size:
    print("All good!")
```

```python
# Create a plot
plt.figure(figsize=(10, 6))

# Plot mass_vector_merged
plt.plot(i, v, label='Datos Experimentales', color='blue', marker='o')

# Adding titles and labels
plt.title('Ley de Ohm')
plt.ylabel('Voltaje [V]')
plt.xlabel('Corriente [A]')

# Show legend
plt.legend()

# Display the plot
plt.grid(True)
plt.show()
```

```python
# Perform linear regression using scipy
slope, intercept, r_value, p_value, std_err = linregress(i, v)

# Generate predictions using the slope and intercept
v_pred = slope * i + intercept

# Plot the regression line
plt.plot(i, v_pred, color='blue', label='label')

# Print the regression statistics
print(f"Slope = {slope:.4f}, Intercept = {intercept:.4f}, R-squared = {r_value**2:.4f}")
```

```python
# Create a plot
plt.figure(figsize=(10, 6))

# Plot experimental data
plt.plot(i, v, label='Datos experimentales', color='blue', marker='o')

# Plot the regression line
plt.plot(i, v_pred, label='Resistencia [9.73 ]', color='red')

# Adding titles and labels
plt.title('Ley de Ohm')
plt.ylabel('Voltaje [V]')
plt.xlabel('Corriente [A]')

# Show legend
plt.legend()

# Display the plot
plt.grid(True)
plt.show()
```

# Osciloscope Signals

```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters for the 1 MHz sine wave
f = 1e6            # Frequency: 1 MHz
T = 1/f            # Period
t = np.linspace(0, 5*T, 1000)  # Plot 5 periods for clarity

# Amplitudes from the experiment
A_ref = 200e-3     # 200 mV for the reference signal (Channel 2)
A_att = 5e-3       # 5 mV for the attenuated signal (Channel 1)

# Generate sine waves (assuming zero phase shift)
signal_ref = A_ref * np.sin(2 * np.pi * f * t)
signal_att = A_att * np.sin(2 * np.pi * f * t)

# Plotting the signals
plt.figure(figsize=(10, 5))
plt.plot(t * 1e6, signal_ref * 1e3, label='Reference Signal (200 mV)', linewidth=2)
plt.plot(t * 1e6, signal_att * 1e3, label='Attenuated Signal (5 mV)', linewidth=2, linestyle='--')
plt.xlabel('Time (s)')
plt.ylabel('Amplitude (mV)')
plt.title('Comparison of 1 MHz Sine Waves')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
