```python
import numpy as np
import matplotlib.pyplot as plt

# Parámetros
rho0 = 1.0             # Valor arbitrario para rho0 (en unidades arbitrarias)
epsilon0 = 8.854e-12   # Permitividad del vacío (F/m)
R = 1.0                # Radio de la esfera (en metros)

# Definición de los valores de r
r = np.linspace(0, 1.5*R, 500)

# Inicializamos el campo eléctrico
E = np.zeros_like(r)

# Para r <= R: E(r) = (rho0/(3*epsilon0))*r*(1 - r/R)
mask = r <= R
E[mask] = (rho0/(3*epsilon0)) * r[mask] * (1 - r[mask]/R)

# Para r > R, según el problema, E(r) = 0 (ya definido)

# Graficamos
plt.figure(figsize=(8, 6))
plt.plot(r, E, label=r'$E(r)=\frac{\rho_0}{3\varepsilon_0}\,r\left(1-\frac{r}{R}\right)$, $r\leq R$')
plt.axvline(R, color='r', linestyle='--', label=r'$r=R$')
plt.xlabel('r (m)')
plt.ylabel('E (N/C)')
plt.title('Magnitud del campo eléctrico vs. r')
plt.legend()
plt.grid(True)
plt.show()
```
