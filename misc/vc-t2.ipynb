```python
import numpy as np
import matplotlib.pyplot as plt

# Número de puntos para muestreo
N = 200
r = np.linspace(1, 2, N)
theta = np.linspace(0, np.pi/2, N)
R, Theta = np.meshgrid(r, theta)
Z = R * np.exp(1j * Theta)

W = Z**3 + 1j  # f(z) = z^3 + i

plt.figure(figsize=(5,4))
plt.scatter(W.real, W.imag, s=1, color='purple')
plt.xlabel('Re(f(z))')
plt.ylabel('Im(f(z))')
plt.title('Imagen de A bajo f(z) = z³ + i')
plt.axis('equal')
plt.grid(True)
plt.show()
```

```python
# Muestreamos para x en (0,3) (para abarcar buena parte del disco) y y en [0,2π]
N = 300
x_vals = np.linspace(0.01, 3, N)  # evitamos x=0
y_vals = np.linspace(0, 2*np.pi, N)
X_vals, Y_vals = np.meshgrid(x_vals, y_vals)
Z = X_vals + 1j * Y_vals

W = np.exp(-np.conjugate(Z))  # f(z) = e^(-conjugado(z))

plt.figure(figsize=(5,4))
plt.scatter(W.real, W.imag, s=1, color='orange')
plt.xlabel('Re(f(z))')
plt.ylabel('Im(f(z))')
plt.title('Imagen de B bajo f(z) = exp(-conjugado(z))')
plt.axis('equal')
plt.grid(True)
plt.show()
```

```python
# Reutilizamos el muestreo de A
N = 200
r = np.linspace(1, 2, N)
theta = np.linspace(0, np.pi/2, N)
R, Theta = np.meshgrid(r, theta)
Z = R * np.exp(1j * Theta)

W = 1/Z  # f(z) = 1/z

plt.figure(figsize=(5,4))
plt.scatter(W.real, W.imag, s=1, color='green')
plt.xlabel('Re(f(z))')
plt.ylabel('Im(f(z))')
plt.title('Imagen de A bajo f(z) = 1/z')
plt.axis('equal')
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Creamos una malla para graficar
x = np.linspace(-3, 3, 400)
y = np.linspace(-3, 3, 400)
X, Y = np.meshgrid(x, y)
Z = X + 1j*Y

### (a) Graficar |iz-1| = |iz+1|, que equivale a y = 0
F = np.abs(1j*Z - 1) - np.abs(1j*Z + 1)

plt.figure(figsize=(5,4))
plt.contour(X, Y, F, levels=[0], colors='blue')
plt.xlabel('Re(z)')
plt.ylabel('Im(z)')
plt.title('Conjunto: |i·z - 1| = |i·z + 1|')
plt.grid(True)
plt.show()

### (b) Graficar la región: Im(z+i) < 2  <=>  y+1 < 2  <=>  y < 1
plt.figure(figsize=(5,4))
plt.contourf(X, Y, Y, levels=[-3, 1], colors=['lightblue'])
plt.axhline(y=1, color='red', linestyle='--', label='y = 1')
plt.xlabel('Re(z)')
plt.ylabel('Im(z)')
plt.title('Conjunto: Im(z+i) < 2')
plt.legend()
plt.grid(True)
plt.show()

### (c) Graficar la franja: 1 <= Re(z) < 2
plt.figure(figsize=(5,4))
plt.contourf(X, Y, X, levels=[1, 2], colors=['lightgreen'])
plt.axvline(x=1, color='red', linestyle='--', label='x = 1')
plt.axvline(x=2, color='red', linestyle='--', label='x = 2')
plt.xlabel('Re(z)')
plt.ylabel('Im(z)')
plt.title('Conjunto: 1 ≤ Re(z) < 2')
plt.legend()
plt.grid(True)
plt.show()
```
